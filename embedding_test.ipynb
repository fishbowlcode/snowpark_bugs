```python
import os
from snowflake.snowpark.session import Session


# Snowflake connection parameters
CONNECTION_PARAMS = {
    "account": os.getenv("SNOWFLAKE_ACCOUNT"),
    "user": os.getenv("SNOWFLAKE_USERNAME"),
    "password": os.getenv("SNOWFLAKE_PASSWORD"),
    "warehouse": os.getenv("SNOWFLAKE_WAREHOUSE"),
    "database": os.getenv("SNOWFLAKE_DATABASE"),
    "schema": os.getenv("SNOWFLAKE_SCHEMA")
}

snowflake_session = Session.builder.configs(
    CONNECTION_PARAMS
).getOrCreate()
```

```python
from langchain_community.embeddings.huggingface import HuggingFaceEmbeddings

# Intialize the Embedding Model
model_name = 'snowflake/snowflake-arctic-embed-l'
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': True}

model = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
from langchain_demo.app.procedures import embed_documents, embed_query

# Define the query and documents
query = "How many dogs go to the park?"
print("Query:", query)

documents = [
    "Dogs go to the park to play.",
    "The park is a popular spot for dogs.",
    "Dogs love to play at the park.",
    "The park is visited by many dogs.",
    "Many dogs go to the park."
]
print("Documents:", documents)

# -- Embed the query --
# Test Procedure Function
embedded_query = embed_query(snowflake_session, query)
# Test Local Model (Above)
# # embedded_query = model.embed_query(query)

# -- Embed the documents --
# Test Procedure Function
embedded_documents = embed_documents(snowflake_session, documents)
# Test Local Model (Above)
# embedded_documents = model.embed_documents(documents)


# Print the embedded query and documents
print("Embedded Query:", embedded_query)
print("Embedded Documents:", embedded_documents)
```
